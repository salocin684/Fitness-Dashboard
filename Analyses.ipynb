# Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import os
import random
```

# Data Distribution

```python
csv_file_path = 'Data\written_name_train_v2.csv'
df = pd.read_csv(csv_file_path, nrows=150000)

random.seed(42)
random_label = random.sample(df['IDENTITY'].unique().tolist(), 5)

label_counts = df['IDENTITY'].value_counts()

random_label_counts = {}
for label in random_label:
    if label in label_counts.index:
        random_label_counts[label] = label_counts[label]
    else:
        random_label_counts[label] = 0

for label, count in random_label_counts.items():
    print(f"Label: {label}, Count: {count}")
```

# Sample Images

```python
image_dir = 'Data/train_v2/train/'

num_rows = 3
num_cols = 4

fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(12,8))

for i, ax in enumerate(axes.flat):
    image_name = f'TRAIN_{i+1:05d}.jpg'
    image_path = os.path.join(image_dir, image_name)
    
    if os.path.exists(image_path):
        img = plt.imread(image_path)
        ax.imshow(img)
        ax.set_title(f'Name: TRAIN_{i+1:04d}')
        ax.axis('off')
        
plt.tight_layout()

plt.show()
```
